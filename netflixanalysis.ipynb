```python
labels=['Movie','TV Show']
explode = [0.1, 0]
counts = df['type'].value_counts().reindex(labels).fillna(0)
plt.pie(counts, labels=labels, autopct='%1.1f%%', explode=explode,startangle=90)
plt.axis('equal')
plt.title('Movies and Shows')
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh')
```

```python
#average release year by type,
df.groupby('type')['release_year'].mean().plot(kind='barh', title='Average Release Per Year')
```
